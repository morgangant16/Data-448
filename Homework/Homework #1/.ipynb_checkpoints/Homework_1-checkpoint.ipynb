```python
import boto3
import pandas as pd
import numpy as np
import xgboost as xgb 
import matplotlib.pyplot as plt
import seaborn as sns


from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, classification_report, precision_recall_curve

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier


s3= boto3.resource('s3')
bucket_name= 'morgan-gant-data448-bucket'
bucket= s3.Bucket(bucket_name)

file_key= 'train.csv'
file_key2= 'test.csv'

bucket_object= bucket.Object(file_key)
file_object = bucket_object.get()
file_content_stream = file_object.get('Body')
bucket_object2= bucket.Object(file_key2)
file_object2 = bucket_object2.get()
file_content_stream2 = file_object2.get('Body')

#reading the datefile
train = pd.read_csv(file_content_stream)
test = pd.read_csv(file_content_stream2)
```

```python
# Defining the input and taregt variables
x= train[['trustLevel', 'scannedLineItemsPerSecond', 'totalScanTimeInSeconds', 'lineItemVoidsPerPosition', 'valuePerSecond']]
y= train['fraud']

#Splitting the data
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size= .2, stratify=y)
```

Model 1: Random Forest

```python
#Creating the model
rf_md= RandomForestClassifier(n_estimators= 500, max_depth=3).fit(x_train, y_train)

#Prediction on the test
rf_pred= rf_md.predict_proba(x_test)[:,1]

#Precision Recall Curve
precision, recall, threshold = precision_recall_curve(y_test, rf_pred)

#Have to take off a percision and recall becasue it has one more than threshold!!
cutoffs= pd.DataFrame({'precision': precision[:-1], 'recall': recall[:-1], 'threshold': threshold})

#Closest distance to perfect model
cutoffs['distance'] = np.sqrt((1 - cutoffs['precision'])**2 + (1 - cutoffs['recall'])**2)

cutoffs= cutoffs.sort_values(by= 'distance').reset_index(drop = True)

# #Chaning the likelyhoods to labels
rf_labels= np.where(rf_pred < cutoffs['threshold'][0], 0, 1)

print(classification_report(y_test, rf_labels))
```

Model 2: XGBoost

```python
#Creating the model
xgb_md= xgb.XGBClassifier(max_depth= 3, n_estimators= 500, learning_rate= .01 ).fit(x_train, y_train)

#predicitng on the model
xgb_pred= xgb_md.predict_proba(x_test)[:,1]
        
#Precision Recall Curve
precision, recall, threshold = precision_recall_curve(y_test, xgb_pred)

#Have to take off a percision and recall becasue it has one more than threshold!!
cutoffs= pd.DataFrame({'precision': precision[:-1], 'recall': recall[:-1], 'threshold': threshold})

#Closest distance to perfect model
cutoffs['distance'] = np.sqrt((1 - cutoffs['precision'])**2 + (1 - cutoffs['recall'])**2)

cutoffs= cutoffs.sort_values(by= 'distance').reset_index(drop = True)

# #Chaning the likelyhoods to labels
xgb_labels= np.where(xgb_pred < cutoffs['threshold'][0], 0, 1)

print(classification_report(y_test, xgb_labels))
  
```
